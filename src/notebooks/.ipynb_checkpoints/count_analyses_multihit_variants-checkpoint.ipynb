# Analyses and figures for total variant count for genes with multiple nonsynonymous hits

```r
# Housekeeping

library(ggplot2)
library(MASS)
library(nlme)
library(scales)
source("source.R")
```

```r
# Read in data

variants = read.table("../../data/deep_seq/multihit_nonsynonymous_variant_data.txt", 
                      sep = "\t", 
                      header = T)
```

```r
# Convert frequency to count data

variants$COUNT = ifelse(variants$VARIANT_FREQUENCY == 0, 0, 1)
```

```r
# Aggregate by species per sample

counts = aggregate(COUNT ~ SPECIES-TIME-ANTIBIOTIC-IMMIGRATION-REPLICATE-AMPLICON_READS, 
                      data = variants, FUN = function(x){sum(x)})

counts = counts[counts$COUNT != 0,] # remove potential 0 count rows
head(counts)
```

```r
# Convert amplicon reads (normalized to 15,000 reads) to relative abundance

counts$REL_ABUNDANCE = counts$AMPLICON_READS/15000

# Log transform (convert zero abundance to extremely low value to prevent downstream problems)

counts$LOG_REL_ABUNDANCE = log10(ifelse(counts$REL_ABUNDANCE == 0,
                              0.0000000000001,
                              counts$REL_ABUNDANCE))
head(counts)
```

```r
# Reformat

counts$SAMPLE = paste(counts$ANTIBIOTIC, 
                      counts$IMMIGRATION, 
                      counts$REPLICATE,
                      sep = "_")
```

## Statistics

### Data exploration

```r
# Balance of categorical covariates

x = as.data.frame(table(counts$SPECIES))
x

# Let's remove species with less than 20 observations

y = x[x$Freq >= 20,]
y
counts2 = counts[counts$SPECIES %in% y$Var1,]

table(counts2$ANTIBIOTIC) # ok
table(counts2$IMMIGRATION) # ok
```

## Model specification

```r
# Let's use generalized least squares models that account for repeated measures

# Define basic model

M1 = gls(COUNT ~ SPECIES + ANTIBIOTIC + IMMIGRATION + TIME + LOG_REL_ABUNDANCE, 
         data = counts2, 
         correlation = corAR1(form = ~ 1 | SAMPLE),
         method = "ML")

# Find best model based on AIC

M2 = stepAIC(M1, 
             scope = list(upper = ~SPECIES*ANTIBIOTIC*IMMIGRATION*TIME*LOG_REL_ABUNDANCE, lower = ~1), 
             trace = FALSE)
```

```r
# Let's also try with species specific variance structure

M3 = gls(COUNT ~ SPECIES + ANTIBIOTIC + IMMIGRATION + TIME + LOG_REL_ABUNDANCE, 
         data = counts2, 
         weights = varIdent(form = ~1 | SPECIES),
         correlation = corAR1(form = ~ 1 | SAMPLE),
         method = "ML")
M4 = stepAIC(M3, 
             scope = list(upper = ~SPECIES*ANTIBIOTIC*IMMIGRATION*TIME*LOG_REL_ABUNDANCE, lower = ~1), 
             trace = FALSE)
```

```r
# Let's compare models with and without species specific variance structure

anova(M2, M4) # M4 better
```

## Model validation

```r
# Homogeneity: residuals vs. predicted values

E1 = resid(M4)
F1 = fitted(M4)
par(mfrow = c(1, 1))
plot(x = F1, 
     y = E1, 
     xlab = "Fitted values",
     ylab = "Residuals")
abline(v = 0, lwd = 2, col = 2)
abline(h = 0, lty = 2, col = 1)
```

```r
# Homogeneity/independence: residuals vs. covariates

plot( x = counts2$SPECIES,
      y = E1,
      xlab = "Species",
      ylab = "Pearson residuals",
      cex.lab = 1.5)
abline(h = 0, lty = 2) # ok

plot( x = counts2$ANTIBIOTIC,
      y = E1,
      xlab = "Antibiotic",
      ylab = "Pearson residuals",
      cex.lab = 1.5)
abline(h = 0, lty = 2) # ok

plot( x = counts2$IMMIGRATION,
      y = E1,
      xlab = "Immigration",
      ylab = "Pearson residuals",
      cex.lab = 1.5)
abline(h = 0, lty = 2) # ok

plot( x = counts2$LOG_REL_ABUNDANCE,
      y = E1,
      xlab = "Relative abundance",
      ylab = "Pearson residuals",
      cex.lab = 1.5)
abline(h = 0, lty = 2) # ok

plot( x = counts2$TIME,
      y = E1,
      xlab = "Time",
      ylab = "Pearson residuals",
      cex.lab = 1.5)
abline(h = 0, lty = 2) # ok
```

```r
# Normality

par(mfrow = c(1, 1))
hist(E1, breaks = 10) # ok
```

## Model inspection

```r
# Model validation plots look ok, let's go with M4

anova(M4)

# species, antibiotic and immigration all explain total variant count
# also all three interact
```

```r
# Species effect

counts$SPECIES = as.factor(as.character(counts$SPECIES))

ggplot(counts, aes(SPECIES, COUNT)) +
    geom_boxplot() +
    ylab("Nonsynonymous mutations in multi-hit genes") +
    coord_flip() +
    scale_x_discrete(limits = rev(levels(counts$SPECIES))) +
    theme(axis.title.y = element_blank())

ggsave("../../manuscript/figures/count_analyses_multihit_variants_species.pdf")
```

```r
# Immigration effect

# Reformat

counts$IMMIGRATION = factor(counts$IMMIGRATION,
                           levels = c("0", "1"),
                           labels = c("No immigration", "Immigration"))
```

```r
# Reformat

counts$ANTIBIOTIC = factor(counts$ANTIBIOTIC, 
                           levels = c("0", "4", "16", "128"),
                           labels = c(expression(paste(0~mu*g*' mL'^-1, " streptomycin")),
                                  expression(paste(4~mu*g*' mL'^-1, " streptomycin")),
                                  expression(paste(16~mu*g*' mL'^-1, " streptomycin")),
                                  expression(paste(128~mu*g*' mL'^-1, " streptomycin"))))
```

```r
# Plot

ggplot(counts, aes(IMMIGRATION, COUNT)) +
    geom_boxplot() +
    ylab("Number of nonsynonymous mutations in multi-hit genes") +
    coord_flip() +
    scale_x_discrete(limits = rev(levels(counts$IMMIGRATION))) +
    facet_grid(~ANTIBIOTIC, labeller = label_parsed) +
    theme(axis.title.y = element_blank())

ggsave("../../manuscript/figures/count_analyses_multihit_variants_immigration_by_antibiotic.pdf", width = 12)
```
