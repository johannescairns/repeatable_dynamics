# Variant frequency distribution plot for filtered variants

```r
# Housekeeping

library(ggplot2)
source("source.R")
```

```r
# Read in data

variants = read.table("../../data/deep_seq/filtered_variant_data.txt", 
                      sep = "\t", 
                      header = T)
```

```r
# Code to make plot prettier

variants$SPECIES = gsub("_", " ", variants$SPECIES)

variants$TREAT = paste(variants$ANTIBIOTIC, variants$IMMIGRATION, variants$TIME, variants$REPLICATE, sep = "_")
a = as.factor(unique(variants$TREAT))
variants$TREAT = factor(variants$TREAT, levels = a)
variants$ABIM = paste(variants$ANTIBIOTIC, variants$IMMIGRATION, sep = "_")
variants$ABIM = factor(variants$ABIM, levels = c("0_0",
                                                 "0_1",
                                                 "4_0",
                                                 "4_1",
                                                 "16_0",
                                                 "16_1",
                                                 "128_0",
                                                 "128_1"),
                       labels = c(expression(atop(paste(0~mu*g*' mL'^-1," Sm"), "No immigration")),
                                  expression(atop(paste(0~mu*g*' mL'^-1," Sm"), "Immigration")),
                                  expression(atop(paste(4~mu*g*' mL'^-1," Sm"), "No immigration")),
                                  expression(atop(paste(4~mu*g*' mL'^-1," Sm"), "Immigration")),
                                  expression(atop(paste(16~mu*g*' mL'^-1," Sm"), "No immigration")),
                                  expression(atop(paste(16~mu*g*' mL'^-1," Sm"), "Immigration")),
                                  expression(atop(paste(128~mu*g*' mL'^-1," Sm"), "No immigration")),
                                  expression(atop(paste(128~mu*g*' mL'^-1," Sm"), "Immigration"))))
```

```r
# Plot code

p1 = ggplot(variants[variants$VARIANT_FREQUENCY > 0,], aes(x = VARIANT_FREQUENCY, fill = SPECIES)) + 
    geom_histogram(bins = 10) +
    facet_grid(SPECIES~ABIM, scales = "free_y",  labeller = label_parsed)+
    scale_fill_manual(values = mycolors) +
    guides(fill = guide_legend(title = "Species")) +
    scale_x_continuous(breaks = seq(0, 1, len = 11)) +
    ylab("Count") +
    xlab("Variant frequency") +
    theme(axis.text.x = element_text(angle = 90, vjust = 0.5),
          strip.text.y = element_blank())
p1
```

```r
# Write out

ggsave("../../manuscript/figures/variant_frequency_distribution.pdf", width = 12, height = 9)
```
