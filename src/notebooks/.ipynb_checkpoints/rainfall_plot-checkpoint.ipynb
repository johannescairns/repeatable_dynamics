# Rainfall plot for filtered variants

```r
# Housekeeping

library(ggplot2)
library(scales)
source("source.R")
```

```r
# Read in data

variants = read.table("../../data/deep_seq/filtered_variant_data.txt", 
                      sep = "\t", 
                      header = T)
```

```r
# Reformat

variants$SPECIES = gsub("_", " ", variants$SPECIES)

# Disinclude first variant in contig (distance to previous variant unknown)

variants = variants[variants$DISTANCE > 0,]

# There is a large number of contigs with varying lengths so normalize to relative position

variants$rel_pos = variants$POSITION / variants$CONTIG_LENGTH
```

```r
# Plot code

p1 = ggplot(variants, aes(y = DISTANCE_FROM_PREV, x = rel_pos, colour = SPECIES)) +
    geom_point() +
    scale_y_log10(labels = trans_format('log10', math_format(10^.x))) +
    scale_colour_manual(values = mycolors) +
    guides(colour = guide_legend(title = "Species")) +
    ylab("Distance to previous variant (log10 bp)") +
    xlab("Relative genomic position within contig")
p1
```

```r
# Write out

ggsave("../../manuscript/figures/rainfall_plot.pdf", width = 8, height = 6)
```
