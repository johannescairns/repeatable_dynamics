# Analyses and figures for allele frequency changes for filtered variants

```r
# Housekeeping

library(car)
library(ggplot2)
library(MASS)
library(mgcv) 
library(nlme)
library(scales)
library(tidyr)
source("source.R")
```

```r
# Read in data

variants = read.table("../../data/deep_seq/filtered_variant_data.txt", 
                      sep = "\t", 
                      header = T)
```

```r
# Convert amplicon reads (normalized to 15,000 reads) to relative abundance

variants$REL_ABUNDANCE = variants$AMPLICON_READS/15000
```

```r
frequency = spread(variants[, 2:20], TIME, VARIANT_FREQUENCY)
colnames(frequency)[18] = "FREQ_T8"
colnames(frequency)[19] = "FREQ_T12"
abundance = spread(variants[, c(2:19, 24)], TIME, REL_ABUNDANCE)
colnames(abundance)[18] = "ABUND_T8"
colnames(abundance)[19] = "ABUND_T12"

variants2 = merge(frequency, abundance, all.x = T)
head(variants2)
```

## Compute selection coefficient and fitness

The slope of a least square regression line (linear model) for logit-transformed allele frequency trajectories (here consisting of two time points) gives the selection coefficient

$s = \frac {logit(f(t_2)) - logit(f(t_1))} {t_2 - t_1}$

where $f(t)$ is the frequency of a mutation at time $t$ (https://www.ncbi.nlm.nih.gov/pubmed/28961717). Fitness $W$ is then given by

$W = s + 1$.

These metrics are used here both for allele frequency changes and changes in the relative abundance of species. In the latter case, they denote the competitive ability of species.

```r
# Compute fitness and selection coefficient for allele frequency changes

# Convert zero abundance alleles to 1 % abundance (detection limit) and fixed alleles to 99.99 % abundance
# to allow estimation of selection coefficients

variants2$FREQ_T8 = ifelse(variants2$FREQ_T8 == 0, 0.01, variants2$FREQ_T8)
variants2$FREQ_T12 = ifelse(variants2$FREQ_T12 == 0, 0.01, variants2$FREQ_T12)

variants2$FREQ_T8 = ifelse(variants2$FREQ_T8 == 1, 0.999, variants2$FREQ_T8)
variants2$FREQ_T12 = ifelse(variants2$FREQ_T12 == 1, 0.999, variants2$FREQ_T12)

# Compute s

variants2$FREQ_S = (logit(variants2$FREQ_T12) - logit(variants2$FREQ_T8))/(12-8)

# Compute W

variants2$FREQ_W = 1 + variants2$FREQ_S

min(variants2$FREQ_S, na.rm = T)
max(variants2$FREQ_S, na.rm = T)
head(variants2)
```

```r
# Compute same metrics for relative abundance changes (competitive ability)

# Convert zero abundance to 10^-6 (order of magnitude lower than lowest detected ABUNDuency)
# to allow estimation of selection coefficients

variants2$ABUND_T8 = ifelse(variants2$ABUND_T8 == 0, 0.000001, variants2$ABUND_T8)
variants2$ABUND_T12 = ifelse(variants2$ABUND_T12 == 0, 0.000001, variants2$ABUND_T12)

# Compute s

variants2$ABUND_S = (logit(variants2$ABUND_T12) - logit(variants2$ABUND_T8))/(12-8)

# Compute W

variants2$ABUND_W = 1 + variants2$ABUND_S

min(variants2$ABUND_S, na.rm = T)
max(variants2$ABUND_S, na.rm = T)
head(variants2)
```

## Statistics

### Data exploration

```r
# Balance of categorical covariates

x = as.data.frame(table(variants2$SPECIES, by = variants2$ANTIBIOTIC))
x

# Let's remove species with no observations for a particular antibiotic concentration

y = as.vector(unique(x$Var1[x$Freq == 0]))
y
variants3 = variants2[!(variants2$SPECIES %in% y),]

table(variants3$ANTIBIOTIC) # ok
table(variants3$IMMIGRATION) # ok

variants3$SPECIES = as.factor(as.character(variants3$SPECIES))
```

```r
# Outliers

par(mfrow = c(1, 2))

# Response variable

boxplot(variants3$FREQ_S, 
        main = "Selection coefficient of variant allele")
dotchart(variants3$FREQ_S, 
         xlab = "Range of data", 
         ylab = "Order of the data")

# Variants seem to separate to three categories (positive, no and negative fitness)

# Covariates

boxplot(variants3$ABUND_S, 
        main = "Competition coefficient of species")
dotchart(variants3$ABUND_S, 
         xlab = "Range of data", 
         ylab = "Order of the data")
```

```r
# missing values

colSums(is.na(variants3))
```

## Model specification

```r
# Start with linear regression

M1 = lm(FREQ_S ~ ANTIBIOTIC + IMMIGRATION + SPECIES + ABUND_S, data = variants3)
M2 = stepAIC(M1,
        scope = list(upper = ~SPECIES*ANTIBIOTIC*IMMIGRATION*ABUND_S, lower = ~1),
        trace = FALSE)
```

```r
# Let's also try with species specific variance structure

M3 = gls(FREQ_S ~ ANTIBIOTIC + IMMIGRATION + SPECIES + ABUND_S, 
         data = variants3, 
         weights = varIdent(form = ~1 | SPECIES),
         method = "ML")
M4 = stepAIC(M3, 
             scope = list(upper = ~SPECIES*ANTIBIOTIC*IMMIGRATION*ABUND_S, lower = ~1), 
             trace = FALSE)
```

```r
# Let's compare models with and without species specific variance structure

AIC(M2, M4) # M4 better
```

## Model validation

```r
# Homogeneity: residuals vs. predicted values

E1 = resid(M4)
F1 = fitted(M4)
par(mfrow = c(1, 1))
plot(x = F1, 
     y = E1, 
     xlab = "Fitted values",
     ylab = "Residuals")
abline(v = 0, lwd = 2, col = 2)
abline(h = 0, lty = 2, col = 1)
```

```r
# Homogeneity/independence: residuals vs. covariates

plot( x = variants3$SPECIES,
      y = E1,
      xlab = "Species",
      ylab = "Pearson residuals",
      cex.lab = 1.5)
abline(h = 0, lty = 2) # ok

plot( x = variants3$ANTIBIOTIC,
      y = E1,
      xlab = "Antibiotic",
      ylab = "Pearson residuals",
      cex.lab = 1.5)
abline(h = 0, lty = 2) # ok

plot( x = variants3$IMMIGRATION,
      y = E1,
      xlab = "Immigration",
      ylab = "Pearson residuals",
      cex.lab = 1.5)
abline(h = 0, lty = 2) # ok

plot( x = variants3$ABUND_S,
      y = E1,
      xlab = "Relative abundance",
      ylab = "Pearson residuals",
      cex.lab = 1.5)
abline(h = 0, lty = 2) # ok
```

```r
# Normality

par(mfrow = c(1, 1))
hist(E1, breaks = 10) # ok
```
