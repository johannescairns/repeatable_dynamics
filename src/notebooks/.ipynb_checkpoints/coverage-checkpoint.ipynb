# Genomic coverage plot

```r
# Housekeeping

library(ggplot2)
library(scales)
library(zoo)
source("source.R")
```

```r
# Read in data

coverage = read.table("../../data/deep_seq/coverage_1kb_window.txt", 
                      sep = "\t", 
                      header = T)
```

```r
head(coverage)
```

```r
# Calculate sliding windows of 10 kb to smoothen data

coverage$rollmean = ave(coverage$reads_per_1k, coverage$SPECIES, 
                         FUN= function(x) rollmean(x, k = 10, na.pad = T))
coverage = na.omit(coverage)
```

```r
head(coverage)
```

```r
# Convert 0 to 0.1 coverage to allow log transformation

coverage$rollmean = ifelse(coverage$rollmean == 0, 0.1, coverage$rollmean)
```

```r
# Code to make plot prettier

#coverage$SPECIES = gsub("_", " ", coverage$SPECIES)

coverage$TREAT = paste(coverage$ANTIBIOTIC, coverage$IMMIGRATION, coverage$TIME, coverage$REPLICATE, sep = "_")
a = as.factor(unique(coverage$TREAT))
coverage$TREAT = factor(coverage$TREAT, levels = a)
coverage$ABIM = paste(coverage$ANTIBIOTIC, coverage$IMMIGRATION, sep = "_")
coverage$ABIM = factor(coverage$ABIM, levels = c("0_0",
                                                 "0_1",
                                                 "4_0",
                                                 "4_1",
                                                 "16_0",
                                                 "16_1",
                                                 "128_0",
                                                 "128_1"),
                       labels = c(expression(atop(paste(0~mu*g*' mL'^-1," Sm"), "No immigration")),
                                  expression(atop(paste(0~mu*g*' mL'^-1," Sm"), "Immigration")),
                                  expression(atop(paste(4~mu*g*' mL'^-1," Sm"), "No immigration")),
                                  expression(atop(paste(4~mu*g*' mL'^-1," Sm"), "Immigration")),
                                  expression(atop(paste(16~mu*g*' mL'^-1," Sm"), "No immigration")),
                                  expression(atop(paste(16~mu*g*' mL'^-1," Sm"), "Immigration")),
                                  expression(atop(paste(128~mu*g*' mL'^-1," Sm"), "No immigration")),
                                  expression(atop(paste(128~mu*g*' mL'^-1," Sm"), "Immigration"))))
```

```r
# Plot code

p1 = ggplot(coverage, aes(x = REL_END, y = rollmean, colour = SPECIES)) + 
    geom_point() +
    facet_grid(ABIM~SPECIES*REPLICATE, scales = "free",  labeller = label_parsed) +
    scale_colour_manual(values = mycolors) +
    guides(colour = guide_legend(title = "Species")) +
    scale_y_log10(label= function(x) {ifelse(x==0, "0", 
                                             parse(text=gsub("[+]", 
                                                             "", 
                                                             gsub("e", " %*% 10^", scientific_format()(x)))))} ) +
    ylab("Coverage") +
    xlab("Relative genomic position") +
    theme(axis.text.x = element_text(angle = 90, vjust = 0.5),
          strip.text.y = element_blank())
p1
```

```r
# Write out

ggsave("../../manuscript/figures/genomic_coverage.pdf", width = 20, height = 20)
```
