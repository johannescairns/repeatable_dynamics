# Random forest models for community composition

Protocol adapted from: 
https://github.com/LangilleLab/microbiome_helper/wiki/Random-Forest-Tutorial (last accessed 06/2019)

**Housekeeping**

```r
library(caret)
library(plyr)
library(randomForest)
library(rfUtilities)
```

**Read in data**

```r
species_composition = read.table("../../../data/amplicon/rel_abund.txt", 
                                 sep = "\t",
                                 header = T,
                                 row.names = 1)

metadata = read.table("../../../data/amplicon/meta.txt",
                      sep = "\t",
                      header = T,
                      row.names = 1)

# inspect

head(species_composition)
head(metadata)
```

**Community resistance (day 32 right after antibiotic pulse)**

Extract data for day 32, transpose and remove missing species

```r
x = metadata$Day == 32
species_composition_d32 = species_composition[x,]
species_composition_d32 = t(species_composition_d32[,colSums(species_composition_d32) > 0])
metadata_d32 = metadata[x,]

# inspect

head(species_composition_d32)
head(metadata_d32)
```

Remove rare features

```r
# plot how many non-zero values each species has

species_nonzero_counts_d32 = apply(species_composition_d32, 1, function(y) sum(length(which(y > 0))))
hist(species_nonzero_counts_d32,
     breaks = 100, 
     col = "grey", 
     main = "", 
     ylab = "Number of species", 
     xlab = "Number of non-zero values")

# R function to remove features that are non-zero less than a specified proportion of the time

remove_rare = function(table, cutoff_pro){
  row2keep = c()
  cutoff = ceiling(cutoff_pro * ncol(table))  
  for (i in 1:nrow(table)){
    row_nonzero = length(which(table[i,] > 0)) 
    if (row_nonzero > cutoff){
      row2keep = c(row2keep, i)
    }
  }
  return(table[row2keep, , drop = F])
}

# discard species that are zero in greater than 80 % of samples

species_table_rare_removed_d32 = remove_rare(table = species_composition_d32, cutoff_pro = 0.2)
                               
# check how many species are left

dim(species_composition_d32)
dim(species_table_rare_removed_d32)

# 17/23 species kept
```

```r
# re-normalize_table so that each sample's column sums to 100 

species_table_rare_removed_norm_d32 = sweep(species_table_rare_removed_d32, 2, 
                                            colSums(species_table_rare_removed_d32), '/') * 100

# inspect

head(colSums(species_table_rare_removed_norm_d32)) # ok
```

Transform data

_Transformations that help make features more comparable across samples can be useful when using relative abundance data. Convert each value into a Z-score by subtracting each sample's mean (center) and then dividing by the sample's standard deviation (scale)._

```r
species_table_scaled_d32 = scale(species_table_rare_removed_norm_d32, center = TRUE, scale = TRUE)  
```

Prep input tables

```r
# classification of antibiotic level

species_table_scaled_d32_AB = data.frame(t(species_table_scaled_d32))
species_table_scaled_d32_AB$AB = metadata[rownames(species_table_scaled_d32_AB), "Streptomycin"]  
species_table_scaled_d32_AB$AB = as.factor(species_table_scaled_d32_AB$AB)

# inspect

head(species_table_scaled_d32_AB)

# classification of presence/absence of immigration

species_table_scaled_d32_IM = data.frame(t(species_table_scaled_d32))  
species_table_scaled_d32_IM$IM = metadata[rownames(species_table_scaled_d32_IM), "Immigration"]  
species_table_scaled_d32_IM$IM = ifelse(species_table_scaled_d32_IM$IM == 0, 1, 2)
species_table_scaled_d32_IM$IM = as.factor(species_table_scaled_d32_IM$IM)

# inspect

head(species_table_scaled_d32_IM)
```

Generate random forest (RF) models

```r
# set random seed for reproducible models

set.seed(151)

# run RF to classify samples based on antibiotic level

RF_AB_d32_classify = randomForest(x = species_table_scaled_d32_AB[,1:(ncol(species_table_scaled_d32_AB)-1)],
                                  y = species_table_scaled_d32_AB[ , ncol(species_table_scaled_d32_AB)],
                                  ntree = 501,
                                  importance = TRUE,
                                  proximities = TRUE)

# run RF to classify samples based on presence/absence of immigration
                                                     
RF_IM_d32_classify = randomForest(x = species_table_scaled_d32_IM[,1:(ncol(species_table_scaled_d32_IM)-1)],
                                  y = species_table_scaled_d32_IM[,ncol(species_table_scaled_d32_IM)],
                                  ntree = 501,
                                  importance = TRUE,
                                  proximities = TRUE )
```

Assess model fit

_If the data had been partitioned into a training and test set beforehand (which was not done in this case due to low sample size) the trained model could be directly validated on independent data. Without this partitioning, the simplest way to assess model fit is to take a look at the model summaries._

```r
RF_AB_d32_classify
RF_IM_d32_classify
```

Permutation test

_One way to assess whether model's performance metric is more extreme than expected by chance is based on a permutation test (here with 1,000 permutations)._

```r
RF_AB_d32_classify_sig = rf.significance(x = RF_AB_d32_classify ,  
                                         xdata = species_table_scaled_d32_AB[,1:(ncol(species_table_scaled_d32_AB)-1)],
                                         nperm = 1000,
                                         ntree = 501)

RF_IM_d32_classify_sig = rf.significance(x = RF_IM_d32_classify,  
                                         xdata = species_table_scaled_d32_IM[,1:(ncol(species_table_scaled_d32_IM)-1)],
                                         nperm = 1000,
                                         ntree = 501)

# inspect

summary(RF_AB_d32_classify_sig)
summary(RF_IM_d32_classify_sig)
```

Accuracy estimated by cross-validation

_One method to estimate model performance is to systematically partition the data into training and test sets and repeatedly see how the model performs (cross-validation). There are many different ways to partition the data, but the simplest is leave-one-out cross-validation. The model is trained n times, where n is the number of samples, and one sample is left out each time for testing. This provides estimates of model performance, which tend to be quite similar to the internal measures (out-of-bag error and mean of squared residuals) described above. We will run this cross-validation using the caret R package. The first step is defining the parameters we want to use for training. In this case we are just specifying leave-one-out cross-validation._

```r
fit_control = trainControl(method = "LOOCV")

# run leave-one-out cross-validation (same ntree and mtry parameters as above)

RF_AB_d32_classify_loocv = train(species_table_scaled_d32_AB[,1:(ncol(species_table_scaled_d32_AB)-1)],
                                 y = species_table_scaled_d32_AB[, ncol(species_table_scaled_d32_AB)],
                                 method = "rf",
                                 ntree = 501,
                                 tuneGrid = data.frame(mtry = 4),
                                 trControl = fit_control)

RF_IM_d32_classify_loocv = train(species_table_scaled_d32_IM[,1:(ncol(species_table_scaled_d32_IM)-1)],
                                 y = species_table_scaled_d32_IM[, ncol(species_table_scaled_d32_IM)],
                                 method = "rf",
                                 ntree = 501 ,
                                 tuneGrid = data.frame(mtry = 4),
                                 trControl = fit_control)

# inspect

RF_AB_d32_classify_loocv$results
RF_IM_d32_classify_loocv$results
```

Identifying important features

```r
# antibiotic level

par(mfrow = c(1,2))

RF_AB_d32_classify_imp = as.data.frame(RF_AB_d32_classify$importance)
RF_AB_d32_classify_imp$features = rownames(RF_AB_d32_classify_imp)
RF_AB_d32_classify_imp_sorted = arrange(RF_AB_d32_classify_imp, desc(MeanDecreaseAccuracy))

# inspect all features

barplot(RF_AB_d32_classify_imp_sorted$MeanDecreaseAccuracy, 
        ylab = "Mean decrease in accuracy (variable importance)", 
        main = "RF classification variable importance distribution")

# inspect top 10 features

barplot(RF_AB_d32_classify_imp_sorted[1:10,"MeanDecreaseAccuracy"], 
        names.arg = RF_AB_d32_classify_imp_sorted[1:10, "features"] , 
        ylab = "Mean decrease in accuracy (variable importance)", 
        las = 2, 
        ylim = c(0,0.02), 
        main = "Classification RF")  

# immigration

RF_IM_d32_classify_imp = as.data.frame(RF_IM_d32_classify$importance)
RF_IM_d32_classify_imp$features = rownames(RF_IM_d32_classify_imp)
RF_IM_d32_classify_imp_sorted = arrange(RF_IM_d32_classify_imp, desc(MeanDecreaseAccuracy))

# inspect all features

barplot(RF_IM_d32_classify_imp_sorted$MeanDecreaseAccuracy, 
        ylab = "Mean decrease in accuracy (variable importance)", 
        main = "RF classification variable importance distribution")

# inspect top 10 features

barplot(RF_IM_d32_classify_imp_sorted[1:10,"MeanDecreaseAccuracy"], 
        names.arg = RF_IM_d32_classify_imp_sorted[1:10, "features"] , 
        ylab = "Mean decrease in accuracy (variable importance)",
        las = 2, 
        ylim = c(0,0.02), 
        main = "Classification RF")
```
