# Phenotype plot

```r
# Housekeeping

library(ggplot2)
library(reshape2)
library(scales)
library(tidyr)
```

```r
# Read in data

phenotypes = read.table("../../../data/amplicon/phenotypic_traits.txt",
                        sep = "\t",
                        header = T)
head(phenotypes)
```

```r
# Plot

ggplot(phenotypes,
       aes(r_growth, Streptomycin_MIC)) +
    geom_point() +
    theme_classic() +
    xlab("Intrinsic growth rate (r)") +
    ylab("Intrinsic streptomycin resistance level (log MIC)")

ggsave("../../../manuscript/figures/growth_rate_vs_resistance_level.pdf", width = 12, height = 10)
```
