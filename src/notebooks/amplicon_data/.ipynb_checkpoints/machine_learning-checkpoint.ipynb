# Random forest (machine learning)

##### Source: https://github.com/LangilleLab/microbiome_helper/wiki/Random-Forest-Tutorial

```r
# Housekeeping

library(caret) # to get leave-one-out cross-validation accuracies and also contains the nearZeroVar function
library(plyr) # for the "arrange" function
library(randomForest)
library(rfUtilities) # to test model significance
```

## Time point 32

```r
# Read in data

species_composition = read.table("../../../data/amplicon/species_composition_relative_abundance.txt", 
                                 sep = "\t",
                                 header = T,
                                 row.names = 1)

metadata = read.table("../../../data/amplicon/metadata.txt",
                      sep = "\t",
                      header = T,
                      row.names = 1)


# Extract regime shift data without predation

x = metadata$Experiment != "FiltrateExp" & # keep only regime shift data
    metadata$Predation != 1 & # exclude predation
    metadata$Immigration != "stock" & # exclude stock
    metadata$Time_point == 32 # keep only day 32

# Subset

species_composition = species_composition[x,] # keep only species with data
species_composition = t(species_composition[,colSums(species_composition)>0])
metadata = metadata[x,-c(3, 6)] # remove redundant columns
# metadata$ID = paste(metadata$Streptomycin, metadata$Immigration, metadata$Replicate, sep = "_")
# metadata$ID = as.factor(metadata$ID)

head(species_composition)
head(metadata)
```

```r
# Pre-processing

# Pre-processing is one of the most important steps in machine learning.
# This is because often the difference between the best performing model isn't the choice of algorithm, 
# but simply how the data is pre-processed. RF doesn't make any assumptions about how the data is distributed 
# so it often is not necessary to transform your data. However, reducing noise in your input data will improve 
# model performance. An easy way to do this is to throw out features that are rare or have very low variance 
# across samples. Any cut-offs used at this step are slightly arbitrary and would depend on the dataset.

# Removing Rare Features

# A basic way to decide which features are unlikely to be informative is if they are non-zero in only a 
# small number of samples. It's a good idea to take a look at this distribution for all features to help 
# choose a reasonable cut-off.

# To quickly figure out and plot how many non-zero values each OTU has:

otu_nonzero_counts <- apply(species_composition, 1, function(y) sum(length(which(y > 0))))
hist(otu_nonzero_counts, breaks=100, col="grey", main="", ylab="Number of OTUs", xlab="Number of Non-Zero Values")
```

```r
# This histogram shows a typical distributions for OTU tables: most OTUs are rare and found in only a few samples. 
# These OTUs are less likely to help our model so we can throw them out. Typically researchers discard OTUs 
# that are zero in greater than 75-90% of samples although these cut-offs are somewhat arbitrary. 
# This cut-off could be optimized if you had an independent dataset or partition of your data that you 
# could use as a validation set.

# This R function removes features that are non-zero less than a specified proportion of the time:

remove_rare <- function( table , cutoff_pro ) {
  row2keep <- c()
  cutoff <- ceiling( cutoff_pro * ncol(table) )  
  for ( i in 1:nrow(table) ) {
    row_nonzero <- length( which( table[ i , ]  > 0 ) ) 
    if ( row_nonzero > cutoff ) {
      row2keep <- c( row2keep , i)
    }
  }
  return( table [ row2keep , , drop=F ])
}

#Based on the above histogram I removed OTUs that have non-zero values in <= 20% of samples:

otu_table_rare_removed <- remove_rare(table=species_composition, cutoff_pro=0.2)
```

```r
# How many OTUs are left?

dim(species_composition)
dim(otu_table_rare_removed)
# 17 species are kept
```

```r
# To re-normalize_table so that each sample's column sums to 100 
# (you can use the ColSums function afterwards for a sanity check):

otu_table_rare_removed_norm <- sweep(otu_table_rare_removed, 2, colSums(otu_table_rare_removed) , '/')*100
colSums(otu_table_rare_removed_norm)

# We might also be interested in excluding OTUs that don't vary much across samples. 
# This usually isn't different from excluding OTUs based on their number of non-zero values. 
# However, this can be useful for other types of feature tables. 
# You may be interested in trying out the nearZeroVar() function from the caret R package to perform this filtering.
```

```r
# Transforming Your Data

# Random forest doesn't make any assumptions about how your input features are distributed 
# so in many cases it isn't necessary to transform your data. However, certain transformations 
# that help make features more comparable across samples can be useful when using relative abundance data.

# One approach is to standardize the data by subtracting each sample's mean (center) and then dividing 
# by the sample's standard deviation (scale). In other words, each value is converted into a Z-score. 
# This is easy to do in R:

otu_table_scaled <- scale(otu_table_rare_removed_norm, center = TRUE, scale = TRUE)  

# Another method is to take the inverse hyperbolic sine and then to mean center by sample, 
# which is also straight-forward to run in R:

otu_table_asinh_mean_centred <- scale( asinh(species_composition), center=TRUE, scale=FALSE)  

# In this case I will use the standardized data although both of these transformations are used in practice. 
# There are many other possible transformations that could be performed as well; you should think carefully 
# about what transformation is most likely to help your model's performance without detracting too much from 
# the interpretability.
```

```r
# Running Model

# Now that we're content with our pre-processing steps, we need to get the input tables into the correct format 
# (note that the metadata column will be the last column of each input dataframe).

# To prep input tables for classification of streptomycin treatment:


otu_table_scaled_SM <- data.frame(t(otu_table_scaled))
otu_table_scaled_SM$SM <- metadata[rownames(otu_table_scaled_SM), "Streptomycin"]  
otu_table_scaled_SM$SM = as.factor(otu_table_scaled_SM$SM)

# To prep input tables for classification of immigration:

otu_table_scaled_IM <- data.frame(t(otu_table_scaled))  
otu_table_scaled_IM$IM <- metadata[rownames(otu_table_scaled_IM), "Immigration"]  
otu_table_scaled_IM$IM = ifelse(otu_table_scaled_IM$IM == 0, 1, 2)
otu_table_scaled_IM$IM = as.factor(otu_table_scaled_IM$IM)

# As described above, the 2 parameters for a RF model are the number of trees in the forest (ntree)
# and the number of features randomly sampled at each node in a tree (mtry). 
# The more trees you run in your forest the better the model will converge. 
# Below I used 501, which is similar to the default, but in practice you may want to use something like 10,001 
# trees for a robust model (depending on the computational time). 
# Note that I usually choose odd numbers of trees to ensure there are never any ties 
# for binary classification models. 
# Unless you have a reason to change mtry beforehand (or can optimize it with an independent partition of data) 
# it's better to use the default values. The default mtry values differ for RF classification and regression - 
# as shown in the below plot.
```

```r
# Finally, before running the models you should set the random seed so that they will be reproducible.

set.seed(151)  

# Run RF to classify samples based on streptomycin treatment:

RF_SM_classify <- randomForest( x=otu_table_scaled_SM[,1:(ncol(otu_table_scaled_SM)-1)] , 
                               y=otu_table_scaled_SM[ , ncol(otu_table_scaled_SM)] , 
                               ntree=501, 
                               importance=TRUE, 
                               proximities=TRUE )

# Run RF to regress OTUs against presence/absence of immigration (IM):
                                                     
RF_IM_classify <- randomForest( x=otu_table_scaled_IM[,1:(ncol(otu_table_scaled_IM)-1)] , 
                              y=otu_table_scaled_IM[ , ncol(otu_table_scaled_IM)] , 
                              ntree=501,
                              importance=TRUE, 
                              proximities=TRUE )
```

```r
# Assessing Model Fit

# Remember that if you partition your data into a training and test set beforehand
# (which we didn't do in this case due to low sample sizes) you would be able to directly validate 
# your trained model on independent data. Without this partitioning the simplest way to assess model 
# fit is to take a look at the model summaries by just typing the model name. 
# For example, for the classification model:

RF_SM_classify
RF_IM_classify
```

```r
# Permutation Test

# Although the above performance metrics provided by the models were encouraging, 
# often it isn't so clear that a model is working well. Testing whether your model's performance metric 
# is more extreme than expected by chance based on a permutation test is one method to assess model significance.

# To run these significance tests with 1000 permutations:

RF_SM_classify_sig <- rf.significance( x=RF_SM_classify ,  xdata=otu_table_scaled_SM[,1:(ncol(otu_table_scaled_SM)-1)] , nperm=1000 , ntree=501 )  
RF_IM_classify_sig <- rf.significance( x=RF_IM_classify ,  xdata=otu_table_scaled_IM[,1:(ncol(otu_table_scaled_IM)-1)] , nperm=1000 , ntree=501 )
summary(RF_SM_classify_sig)
summary(RF_IM_classify_sig)
```

```r
# Accuracy Estimated by Cross-validation

# One method to estimate model performance is to systematically partition the data into training 
# and test sets and repeatedly see how the model performs (cross-validation). There are many different ways to partition the data, but the simplest is leave-one-out cross-validation. The model is trained n times, where n is the number of samples, and one sample is left out each time for testing. This provides estimates of model performance, which tend to be quite similar to the internal measures (out-of-bag error and mean of squared residuals) described above. We will run this cross-validation using the caret R package.

# The first step is defining the parameters we want to use for training:

fit_control <- trainControl( method = "LOOCV" )    

# In this case we are just specifying leave-one-out cross-validation, but caret can be used for 
# much more sophisticated purposes.

# These commands would run the leave-one-out cross-validation, note the same ntree and mtry parameters 
# are being used as above:

RF_SM_classify_loocv <- train( otu_table_scaled_SM[,1:(ncol(otu_table_scaled_SM)-1)] ,
                              y=otu_table_scaled_SM[, ncol(otu_table_scaled_SM)] , 
                              method="rf", 
                              ntree=501 , 
                              tuneGrid=data.frame( mtry=4 ) , 
                              trControl=fit_control )

RF_IM_classify_loocv <- train( otu_table_scaled_IM[,1:(ncol(otu_table_scaled_IM)-1)] ,
                              y=otu_table_scaled_IM[, ncol(otu_table_scaled_IM)] , 
                              method="rf", 
                              ntree=501 , 
                              tuneGrid=data.frame( mtry=4 ) , 
                              trControl=fit_control )
```

```r
# To see the performance metrics:

RF_SM_classify_loocv$results
RF_IM_classify_loocv$results

# Interpretation for randomForest example data set (not these results):

# Both the Accuracy and Kappa metrics are 100%, which should make us extremely confident in our classification 
# model (you won't see models perform this well usually!). Also, the R-squared of the regression model 
# is very similar to what we previously found based on the internal testing, which is typical since the 
# internal RF performance metrics should not be biased.
```

```r
# Identifying Important Features

par(mfrow=c(1,2))
RF_SM_classify_imp <- as.data.frame( RF_SM_classify$importance )
RF_SM_classify_imp$features <- rownames( RF_SM_classify_imp )
RF_SM_classify_imp_sorted <- arrange( RF_SM_classify_imp  , desc(MeanDecreaseAccuracy)  )
barplot(RF_SM_classify_imp_sorted$MeanDecreaseAccuracy, ylab="Mean Decrease in Accuracy (Variable Importance)", 
        main="RF Classification Variable Importance Distribution")

RF_IM_classify_imp <- as.data.frame( RF_IM_classify$importance )
RF_IM_classify_imp$features <- rownames( RF_IM_classify_imp )
RF_IM_classify_imp_sorted <- arrange( RF_IM_classify_imp  , desc(MeanDecreaseAccuracy)  )
barplot(RF_IM_classify_imp_sorted$MeanDecreaseAccuracy, ylab="Mean Decrease in Accuracy (Variable Importance)", 
        main="RF Classification Variable Importance Distribution")

# Often it's more informative to look at a subset of top features, for instance the top 10 features:

barplot(RF_SM_classify_imp_sorted[1:10,"MeanDecreaseAccuracy"], 
        names.arg=RF_SM_classify_imp_sorted[1:10,"features"] , 
        ylab="Mean Decrease in Accuracy (Variable Importance)", las=2, ylim=c(0,0.02), main="Classification RF")  

barplot(RF_IM_classify_imp_sorted[1:10,"MeanDecreaseAccuracy"], 
        names.arg=RF_IM_classify_imp_sorted[1:10,"features"] , 
        ylab="Mean Decrease in Accuracy (Variable Importance)", las=2, ylim=c(0,0.02), main="Classification RF")  
```

### Interpretation for time point 32

#### Reasonably good classification obtained for streptomycin levels, with some overlap between no antibiotic and lowest level. Three species (1972, 1287 and 1896) are particularly important for outcome. Immigration cannot be classified well.
