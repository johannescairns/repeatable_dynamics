# Community size analyses

```r
# Housekeeping

library(emmeans)
library(ggplot2)
library(MASS)
library(nlme)
library(Rmisc)
library(scales)
```

```r
# Read in data

community_size = read.table("../../../data/amplicon/community_size.txt",
                            sep = "\t",
                            header = T)

# Extract no ciliate data

bacteria = community_size[community_size$Predation == 0, 
                                c("Time_point", "Streptomycin", "Immigration", "Replicate", "OD600")]
bacteria$Time_point = gsub("4", 16, bacteria$Time_point)
bacteria$Time_point = gsub("8", 32, bacteria$Time_point)
bacteria$Time_point = gsub("12", 48, bacteria$Time_point)
head(bacteria)
```

```r
# Bacterial population size

# Add sample id

bacteria$Sample = paste(bacteria$Streptomycin, bacteria$Immigration, bacteria$Replicate, sep = "_")

# Let's use generalized least squares models that account for repeated measures

# Define basic model

M1 = gls(OD600 ~ Streptomycin + Immigration + Time_point, 
         data = bacteria, 
         correlation = corAR1(form = ~ 1 | Sample),
         method = "ML")

# Find best model based on AIC

M2 = stepAIC(M1, 
             scope = list(upper = ~Streptomycin * Immigration * Time_point, lower = ~1), 
             trace = FALSE)
```

```r
# Let's also try with streptomycin dependent variance structure

M3 = gls(OD600 ~ Streptomycin + Immigration + Time_point, 
         data = bacteria, 
         weights = varIdent(form = ~1 | Streptomycin),
         correlation = corAR1(form = ~ 1 | Sample),
         method = "ML")

#M4 = stepAIC(M3, 
#             scope = list(upper = ~Streptomycin * Immigration, lower = ~1), 
#             trace = FALSE)
```

```r
# Let's compare models with and without species specific variance structure

anova(M2, M3) # M2 better
```

## Model validation

```r
# Homogeneity: residuals vs. predicted values

E1 = resid(M2)
F1 = fitted(M2)
par(mfrow = c(1, 1))
plot(x = F1, 
     y = E1, 
     xlab = "Fitted values",
     ylab = "Residuals")
abline(v = 0, lwd = 2, col = 2)
abline(h = 0, lty = 2, col = 1)
```

```r
# Homogeneity/independence: residuals vs. covariates

plot( x = bacteria$Streptomycin,
      y = E1,
      xlab = "Streptomycin",
      ylab = "Pearson residuals",
      cex.lab = 1.5)
abline(h = 0, lty = 2) # ok

plot( x = bacteria$Immigration,
      y = E1,
      xlab = "Immigration",
      ylab = "Pearson residuals",
      cex.lab = 1.5)
abline(h = 0, lty = 2) # ok

plot( x = bacteria$Time_point,
      y = E1,
      xlab = "Time",
      ylab = "Pearson residuals",
      cex.lab = 1.5)
abline(h = 0, lty = 2) # ok
```

```r
# Normality

par(mfrow = c(1, 1))
hist(E1, breaks = 10) # ok
```

## Model inspection

```r
# Model validation plots look ok, let's go with M4

anova(M2)

# Streptomycin and time explain bacterial population size
```
