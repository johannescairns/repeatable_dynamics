# Replicator equation to assess competitive fitness of species

```r
# Housekeeping

library(car)
library(MASS)
library(nlme)
library(reshape2)
library(scales)
library(tidyverse)
```

```r
# Read in data

species_composition = read.table("../../../data/amplicon/species_composition_relative_abundance.txt", 
                                 sep = "\t",
                                 header = T,
                                 row.names = 1)

metadata = read.table("../../../data/amplicon/metadata.txt",
                      sep = "\t",
                      header = T,
                      row.names = 1)


# Extract regime shift data without predation

x = metadata$Experiment != "FiltrateExp" & # keep only regime shift data
    metadata$Predation != 1 & # exclude predation
    metadata$Immigration != "stock" # exclude stock

# Subset

species_composition = species_composition[x,] # keep only species with data
species_composition = species_composition[,colSums(species_composition)>0]
metadata = metadata[x,-c(3, 6)] # remove redundant columns
#species_composition = as.data.frame(ifelse(species_composition == 0, 0.0000000000001, species_composition))

species_composition$SAMPLE = rownames(species_composition)
metadata$SAMPLE = rownames(metadata)

df = merge(species_composition, metadata, all = T)

dim(df)
head(df)
```

```r
df2 = melt(df[,-1], id.vars = c("Replicate", "Streptomycin", "Immigration", "Time_point"))
colnames(df2) = c("Replicate", "Streptomycin", "Immigration", "Time_point", "Species", "Abundance")
head(df2)
```

```r
df3 = spread(df2, Time_point, Abundance)
colnames(df3) = c("Replicate", "Streptomycin", "Immigration", "Species", "ABUND16", "ABUND32", "ABUND48")
df3$ID = paste(df3$Streptomycin, df3$Immigration, df3$Replicate, sep = "_")
head(df3)
```

```r
# Compute selection coefficients leveraged against initial population size

selcoef1 = df3[df3$ABUND16 > 0 | df3$ABUND32 > 0,]
selcoef1$ABUND16 = ifelse(selcoef1$ABUND16 < 0.000001, 0.000001, selcoef1$ABUND16)
selcoef1$ABUND32 = ifelse(selcoef1$ABUND32 < 0.000001, 0.000001, selcoef1$ABUND32)
selcoef1$selcoef = (selcoef1$ABUND32-selcoef1$ABUND16)/selcoef1$ABUND16
selcoef1$Interval = "pulse"

selcoef2 = df3[df3$ABUND32 > 0 | df3$ABUND48 > 0,]
selcoef2$ABUND32 = ifelse(selcoef2$ABUND32 < 0.000001, 0.000001, selcoef2$ABUND32)
selcoef2$ABUND48 = ifelse(selcoef2$ABUND48 < 0.000001, 0.000001, selcoef2$ABUND48)
selcoef2$selcoef = (selcoef2$ABUND48-selcoef2$ABUND32)/selcoef2$ABUND32
selcoef2$Interval = "recovery"

df4 = rbind(selcoef1, selcoef2)
head(df4)
```

```r
# Log transform fitness (several very low values skew distribution)

df4$selcoef_log = log10(df4$selcoef + 1 - min(df4$selcoef)) # translate, then transform
min(df4$selcoef_log)
max(df4$selcoef_log)
head(df4)
```

Add phenotype data and see if results are comparable

```r
# Incorporate phenotypes

phenotypes = read.table("../../../data/amplicon/phenotypic_traits.txt",
                        sep = "\t",
                        header = T)

phenotypes$Species = rownames(phenotypes)
dim(phenotypes)
phenotypes = phenotypes[phenotypes$Species %in% unique(df4$Species),]
dim(phenotypes)
df5 = merge(df4, phenotypes, by = "Species", all = T)
dim(df4)
dim(df5)
head(df5)
```

```r
# Add sample id

df5$Sample = paste(df5$Streptomycin, df5$Immigration, df5$Replicate, sep = "_")
```

```r
# Stats

M1 = gls(selcoef_log ~ factor(Streptomycin) + Immigration + Streptomycin_MIC + r_growth + Interval, 
         data = df5,
         weights = varIdent(form = ~1 | factor(Streptomycin)),
         correlation = corAR1(form = ~ 1 | Sample),
         method = "ML")

M2 = stepAIC(M1, 
             scope = list(upper = ~factor(Streptomycin) * Immigration * Streptomycin_MIC * r_growth*Interval, lower = ~1), 
             trace = FALSE)
```

```r
variants2 = spread(variants[, 2:20], TIME, VARIANT_FREQUENCY)
colnames(variants2)[18] = "FREQ_T8"
colnames(variants2)[19] = "FREQ_T12"
```

## Compute selection coefficient and fitness

The slope of a least square regression line (linear model) for logit-transformed allele frequency trajectories (here consisting of two time points) gives the selection coefficient

$s = \frac {logit(f(t_2)) - logit(f(t_1))} {t_2 - t_1}$

where $f(t)$ is the frequency of a mutation at time $t$ (https://www.ncbi.nlm.nih.gov/pubmed/28961717). Fitness $W$ is then given by

$W = s + 1$.

```r
# Compute fitness and selection coefficient for allele frequency changes

# Convert zero abundance alleles to 1 % abundance (detection limit) and fixed alleles to 99.99 % abundance
# to allow estimation of selection coefficients

variants2$FREQ_T8 = ifelse(variants2$FREQ_T8 == 0, 0.01, variants2$FREQ_T8)
variants2$FREQ_T12 = ifelse(variants2$FREQ_T12 == 0, 0.01, variants2$FREQ_T12)

variants2$FREQ_T8 = ifelse(variants2$FREQ_T8 == 1, 0.999, variants2$FREQ_T8)
variants2$FREQ_T12 = ifelse(variants2$FREQ_T12 == 1, 0.999, variants2$FREQ_T12)

# Compute s

variants2$FREQ_S = (logit(variants2$FREQ_T12) - logit(variants2$FREQ_T8))/(12-8)

variants2 = variants2[!is.na(variants2$FREQ_S),]

# Compute W

variants2$FREQ_W = 1 + variants2$FREQ_S

dim(variants2)
head(variants2)
```

```r
# Start with linear regression

M3 = lm(FREQ_S ~ ANTIBIOTIC + IMMIGRATION + SPECIES + selcoef_log_recovery, data = variants4)
M4 = stepAIC(M3,
        scope = list(upper = ~SPECIES*ANTIBIOTIC*IMMIGRATION*selcoef_log_recovery, lower = ~1),
        trace = FALSE)

# Let's also try with species specific variance structure

M5 = gls(FREQ_S ~ ANTIBIOTIC + IMMIGRATION + SPECIES + selcoef_log_recovery, 
         data = variants4, 
         weights = varIdent(form = ~1 | SPECIES),
         method = "ML")
M6 = stepAIC(M5, 
             scope = list(upper = ~SPECIES*ANTIBIOTIC*IMMIGRATION*selcoef_log_recovery, lower = ~1), 
             trace = FALSE)

# Let's compare models with and without species specific variance structure

AIC(M4, M6) # M4 better
```

```r
# Normalize fitness values to 0-1 range and summarize

df8 = dcast(df5[df5$Interval == "pulse",], Species ~ Streptomycin*Immigration*Replicate, value.var = "selcoef")
library(vegan)
rownames(df8) = df8$Species
#df8 = decostand(df8[,-1], method = "range", na.rm = T)
df8$Species = rownames(df8)
df8 = melt(df8, id.vars = "Species")
df8[is.na(df8)] = 0
df8$Streptomycin = gsub("_.*", "", df8$variable)
df8 = df8[,-2]
colnames(df8) = c("Species", "Fitness", "Streptomycin")
df8 = summarySE(df8, measurevar = "Fitness", groupvars = c("Species", "Streptomycin"))
df8[is.na(df8)] = 0
head(df8)
```

```r
df8$Species = as.factor(df8$Species)
df8$Streptomycin = as.numeric(as.character(df8$Streptomycin))
df8$Streptomycin = ifelse(df8$Streptomycin == 0, 0, ifelse(df8$Streptomycin == 4, 1, ifelse(df8$Streptomycin == 16, 2, 3)))
```

```r
dim(df8)
df9 = df5[,c("Species", "MIC_orig", "r_growth")]
df9 = df9[!duplicated(df9),]
dim(df9)
df10 = merge(df8, df9, by = "Species", all = T)
dim(df8)
dim(df10)
head(df10)
```

```r
df10$Species = gsub(".*_HAMBI", "HAMBI", df10$Species)
df10$Species = gsub(".*_JE", "JE", df10$Species)
df10$Species = gsub("_RP4", "", df10$Species)
df10$Species = gsub("_", " ", df10$Species)
species_levels = unique(df10$Species)
species_levels = species_levels[order(as.numeric(as.character(sub('.* ', '', species_levels))))]
df10$Species = factor(df10$Species, levels = species_levels)
```

```r
df10$r_quant = ifelse(df10$r_growth <= quantile(df10$r_growth, 0.33), 1, 
                      ifelse(df10$r_growth <= quantile(df10$r_growth, 0.66), 2, 3))
df10$MIC_quant = ifelse(df10$MIC_orig <= quantile(df10$MIC_orig, 0.33), 1, 
                      ifelse(df10$MIC_orig <= quantile(df10$MIC_orig, 0.66), 2, 3))
df10$quant = paste(df10$r_quant, df10$MIC_quant, sep = "_")
```

```r
unique(df10$quant)
df10$quant = gsub("1_1", "Low growth rate, low resistance level", df10$quant)
df10$quant = gsub("1_2", "Low growth rate, intermediate resistance level", df10$quant)
df10$quant = gsub("1_3", "Low growth rate, high resistance level", df10$quant)
df10$quant = gsub("2_1", "Intermediate growth rate, low resistance level", df10$quant)
df10$quant = gsub("2_2", "Intermediate growth rate, intermediate resistance level", df10$quant)
df10$quant = gsub("2_3", "Intermediate growth rate, high resistance level", df10$quant)
df10$quant = gsub("3_1", "High growth rate, low resistance level", df10$quant)
df10$quant = gsub("3_2", "High growth rate, intermediate resistance level", df10$quant)
df10$quant = gsub("3_3", "High growth rate, high resistance level", df10$quant)
df10$quant = factor(df10$quant, levels = c("Low growth rate, low resistance level",
                                           "Low growth rate, intermediate resistance level",
                                           "Low growth rate, high resistance level",
                                           "Intermediate growth rate, low resistance level",
                                           "Intermediate growth rate, intermediate resistance level",
                                           "Intermediate growth rate, high resistance level",
                                           "High growth rate, low resistance level",
                                           "High growth rate, intermediate resistance level",
                                           "High growth rate, high resistance level"))
head(df10)
```
