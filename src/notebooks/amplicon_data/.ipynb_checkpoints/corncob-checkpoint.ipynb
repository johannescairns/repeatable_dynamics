# Differential abundance and variability analysis using a new beta-binomial regression approach (corncob)

### The approach models each taxon individually

### https://arxiv.org/abs/1902.02776

```r
# Housekeeping

library(ggplot2)
library(phyloseq)

# devtools::install_github("bryandmartin/corncob")
library(corncob)

library(reshape2)
library(tidyr)
```

```r
# Read in data

species_composition = read.table("../../../data/amplicon/species_composition_normalized_to_15k_reads.txt", 
                                 sep = "\t",
                                 header = T,
                                 row.names = 1)

metadata = read.table("../../../data/amplicon/metadata.txt",
                      sep = "\t",
                      header = T,
                      row.names = 1)

head(species_composition)
head(metadata)
```

```r
# Extract regime shift data without predation

x = metadata$Experiment == "RegimeShift" & # keep only regime shift data
    metadata$Predation == 0 & # exclude predation
    metadata$Immigration != "stock" # exclude initial strain mix data

# Subset

species_composition = species_composition[x,]
metadata = metadata[x,]

head(species_composition)
head(metadata)
```

## Differential abundance day 32

```r
# Create phyloseq object for day 32

x = metadata$Time_point == 32
species_composition_T32 = species_composition[x,]
metadata_T32 = metadata[x,]

# Species abundance data

otumat = t(species_composition_T32[,colSums(species_composition_T32) > 0]) # remove missing species
OTU = otu_table(otumat, taxa_are_rows = TRUE)

# Taxonomic information

taxmat = matrix(sample(letters, nrow(otumat)*7, replace = TRUE), nrow = nrow(otumat), ncol = 7)
rownames(taxmat) <- rownames(otumat)
colnames(taxmat) <- c("Domain", "Phylum", "Class", "Order", "Family", "Genus", "Species")
taxmat = as.data.frame(taxmat)
taxmat$Genus = vapply(strsplit(row.names(taxmat), "_", fixed = T), "[", "", 1)
taxmat$Species = vapply(strsplit(row.names(taxmat), "_", fixed = T), "[", "", 2)
taxmat = as.matrix(taxmat)
TAX = tax_table(taxmat)

# Sample metadata including time point

sampledata = sample_data(metadata_T32[, -c(3, 5:6)]) # remove redundant columns

# Combine data types into phyloseq object

physeq_T32 = phyloseq(OTU, TAX, sampledata)
physeq_T32
```

```r
da_analysis_T32 = differentialTest(formula = ~ factor(Immigration) * factor(Streptomycin),
                                   phi.formula = ~ factor(Immigration) * factor(Streptomycin),
                                   formula_null = ~ 1,
                                   phi.formula_null = ~ factor(Immigration) * factor(Streptomycin),
                                   test = "Wald",
                                   boot = FALSE,
                                   data = physeq_T32,
                                   fdr_cutoff = 0.05)
plot(da_analysis_T32)
```

## Differential abundance day 48

```r
# Create phyloseq object for day 48

x = metadata$Time_point == 48
species_composition_T48 = species_composition[x,]
metadata_T48 = metadata[x,]

# Species abundance data

otumat = t(species_composition_T48[,colSums(species_composition_T48) > 0]) # remove missing species
OTU = otu_table(otumat, taxa_are_rows = TRUE)

# Taxonomic information

taxmat = matrix(sample(letters, nrow(otumat)*7, replace = TRUE), nrow = nrow(otumat), ncol = 7)
rownames(taxmat) <- rownames(otumat)
colnames(taxmat) <- c("Domain", "Phylum", "Class", "Order", "Family", "Genus", "Species")
taxmat = as.data.frame(taxmat)
taxmat$Genus = vapply(strsplit(row.names(taxmat), "_", fixed = T), "[", "", 1)
taxmat$Species = vapply(strsplit(row.names(taxmat), "_", fixed = T), "[", "", 2)
taxmat = as.matrix(taxmat)
TAX = tax_table(taxmat)

# Sample metadata including time point

sampledata = sample_data(metadata_T48[, -c(3, 5:6)]) # remove redundant columns

# Combine data types into phyloseq object

physeq_T48 = phyloseq(OTU, TAX, sampledata)
physeq_T48
```

```r
da_analysis_T48 = differentialTest(formula = ~ factor(Immigration) * factor(Streptomycin),
                                   phi.formula = ~ factor(Immigration) * factor(Streptomycin),
                                   formula_null = ~ 1,
                                   phi.formula_null = ~ factor(Immigration) * factor(Streptomycin),
                                   test = "Wald",
                                   boot = FALSE,
                                   data = physeq_T32,
                                   fdr_cutoff = 0.05)
plot(da_analysis_T48)
```

```r
da_analysis_T32$significant_taxa == da_analysis_T48$significant_taxa

# Same species differentially abundant / variable in both cases
```

```r
# Edit data for plotting

species_composition_diff_ab = species_composition[,colnames(species_composition) %in% 
                                                  da_analysis_T48$significant_taxa]
species_composition_diff_ab$label = rownames(species_composition_diff_ab)
metadata$label = rownames(metadata)
diff_ab = merge(metadata[,-c(3,6)], species_composition_diff_ab, all = T)
diff_ab = melt(diff_ab[,-1], id.vars = c("Replicate", "Streptomycin", "Immigration", "Time_point"))
colnames(diff_ab) = c("Replicate", "Streptomycin", "Immigration", "Time_point", "Species", "Normalized_read_count")
head(diff_ab)
```

```r
# Plot differentially abundant species

ggplot(diff_ab, aes(Time_point, Normalized_read_count, colour = factor(Streptomycin))) +
    geom_point() +
    geom_line() +
    facet_grid(Species~Immigration*Replicate, scales = "free") +
    scale_color_manual(values = c("#D3D3D3", "#cd6090", "#8f4364", "#522639")) + 
    theme_bw() +
    ylab("Read count normalized to 15k reads per community") +
    labs(colour = "Streptomycin level") +
    theme(strip.text.y = element_text(angle = 0),
        legend.direction = "horizontal", 
        legend.position = "bottom",
        legend.box = "horizontal",
        strip.background = element_rect(colour = "white", fill = "white"),
        strip.text = element_text(face = "italic", colour = "black"),
        panel.border = element_rect(colour = "grey"),
        panel.grid.major = element_blank(),
        panel.grid.minor = element_blank())
```

### Example for individual species

```r
# Produce baseline model

M1_T32_ac = bbdml(formula = Aeromonas_caviae_HAMBI_1972 ~ factor(Streptomycin) * factor(Immigration),
                  phi.formula = ~ factor(Streptomycin) * factor(Immigration),
                  data = physeq_T32) 
```

```r
# Drop streptomycin*immigration interaction

M2_T32_ac = bbdml(formula = Aeromonas_caviae_HAMBI_1972 ~ factor(Streptomycin) + factor(Immigration),
                  phi.formula = ~ factor(Streptomycin) + factor(Immigration),
                  data = physeq_T32) 
```

```r
# Compare models

waldchisq(M1_T32_ac, M2_T32_ac) # interaction NS (p = 0.41) --> continue with simpler model M2_T32_ac
```

```r
# Drop streptomycin

M3_T32_ac = bbdml(formula = Aeromonas_caviae_HAMBI_1972 ~ factor(Immigration),
                  phi.formula = ~ factor(Immigration),
                  data = physeq_T32)
```

```r
# Compare models

waldchisq(M2_T32_ac, M3_T32_ac) # sig. sm effect (p < 0.001) --> continue with M2_T32_ac
```

```r
# Drop immigration

M4_T32_ac = bbdml(formula = Aeromonas_caviae_HAMBI_1972 ~ factor(Streptomycin),
                  phi.formula = ~ factor(Streptomycin),
                  data = physeq_T32)
```

```r
# Compare models

waldchisq(M2_T32_ac, M4_T32_ac) # sig. immigration effect (p < 0.001) --> continue with M2_T32_ac
```

```r
# Inspect model

M2_T32_ac

# abundance differs from no antibiotic in all antibiotic concentrations
# abundance differs with immigration
# variability does not differ with antibiotic or immigration
```
