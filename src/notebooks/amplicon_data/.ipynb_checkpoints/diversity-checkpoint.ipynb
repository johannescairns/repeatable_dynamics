# Diversity analysis

```r
# Housekeeping

library(ggplot2)
library(lme4)
library(multcomp)
library(reshape2)
library(scales)
library(vegan)
source("../source.R")
```

```r
# Read in data

species_composition = read.table("../../../data/amplicon/species_composition_normalized_to_15k_reads.txt", 
                                 sep = "\t",
                                 header = T,
                                 row.names = 1)

metadata = read.table("../../../data/amplicon/metadata.txt",
                      sep = "\t",
                      header = T,
                      row.names = 1)

head(species_composition)
head(metadata)
```

```r
# Compute diversity (inverse Simpson index)

diversity = diversity(species_composition, "shannon")
head(diversity)
```

```r
# Merge with metadata

diversity = data.frame(Diversity = diversity, metadata)
head(diversity)
```

## (1) Regime shift

```r
# Subset regime shift data

# Select data (rows) to keep

diversity_regime_shift = diversity[diversity$Experiment != "FiltrateExp",]
head(diversity_regime_shift)
```

### (a) Without predation

```r
# Subset predation-free data

diversity_regime_shift_nop = diversity_regime_shift[diversity_regime_shift$Predation == 0,]
head(diversity_regime_shift_nop)
```

#### Prior to antibiotic pulse

```r
# Subset time to before antibiotic pulse

diversity_regime_shift_nop_before = diversity_regime_shift_nop[diversity_regime_shift_nop$Time_point == 16,]
head(diversity_regime_shift_nop_before)
```

```r
# Test for immigration effect

M1 = lm(Diversity ~ Immigration, data = diversity_regime_shift_nop_before)
anova(M1) # no effect of imigration before antibiotic addition
```

#### Resistance

```r
# Subset time to immediately after antibiotic pulse

diversity_regime_shift_nop_resistance = diversity_regime_shift_nop[diversity_regime_shift_nop$Time_point == 32,]
head(diversity_regime_shift_nop_resistance)
```

```r
# Test for treatment effects

diversity_regime_shift_nop_resistance$Streptomycin = as.factor(diversity_regime_shift_nop_resistance$Streptomycin)
M2 = lm(Diversity ~ Streptomycin * Immigration, data = diversity_regime_shift_nop_resistance)
anova(M2) # streptomycin, immigration and interaction effects
```

```r
# Post hoc test for streptomycin effect

summary(glht(M2, linfct = mcp(Streptomycin = 'Tukey', interaction_average = T)))
# All pairwise differences significant except for 16 vs. 128 ug/ml streptomycin
```

```r
# Define color palette

mypalette = c("#D3D3D3", "#cd6090", "#8f4364", "#522639")
```

```r
# Visualize effect

diversity_regime_shift_nop_resistance$Immigration = as.factor(ifelse(diversity_regime_shift_nop_resistance$Immigration == 0, "Absent", "Present"))

ggplot(diversity_regime_shift_nop_resistance, aes(x = Immigration, y = Diversity, colour = Streptomycin)) +
    geom_boxplot(outlier.shape = NA) +
    geom_point(position = position_jitterdodge()) +
    scale_colour_manual(values = mypalette) +
    xlab("Immigration") +
    ylab("Diversity (inverse Simpson index)") +
    labs(colour = "Streptomycin level") +
    theme_classic()
```

##### Summary for resistance without predation: Diversity decreases with increasing antibiotic concentration until 16 ug/ml. Diversity also decreases without streptomycin. Immigration counteracts this but only for streptomycin-free and lowest concentration treatments.

#### Resistance

```r
# Subset time to after recovery from antibiotic

diversity_regime_shift_nop_resilience = diversity_regime_shift_nop[diversity_regime_shift_nop$Time_point == 48,]
head(diversity_regime_shift_nop_resilience)
```

```r
# Test for immigration effect

diversity_regime_shift_nop_resilience$Streptomycin = as.factor(diversity_regime_shift_nop_resilience$Streptomycin)
M3 = lm(Diversity ~ Streptomycin * Immigration, data = diversity_regime_shift_nop_resilience)
anova(M3) # streptomycin, immigration and interaction effects
```

```r
# Post hoc test for streptomycin effect

summary(glht(M3, linfct = mcp(Streptomycin = 'Tukey', interaction_average = T)))
# Difficult to interpret because of interaction, let's visualize
```

```r
# Visualize effect

diversity_regime_shift_nop_resilience$Immigration = as.factor(ifelse(diversity_regime_shift_nop_resilience$Immigration == 0, "Absent", "Present"))

ggplot(diversity_regime_shift_nop_resilience, aes(x = Immigration, y = Diversity, colour = Streptomycin)) +
    geom_boxplot(outlier.shape = NA) +
    geom_point(position = position_jitterdodge()) +
    scale_colour_manual(values = mypalette) +
    xlab("Immigration") +
    ylab("Diversity (inverse Simpson index)") +
    labs(colour = "Streptomycin level") +
    theme_classic()
```

##### Summary for resilience without predation: Without immigration, diversity increases at low to moderate while decreasing at high disturbance. When immigration and disturbance are combined, diversity increases.

## (2) Filtrate experiment

```r
# Subset regime shift data

# Select data (rows) to keep

diversity_filtrate = diversity[diversity$Experiment == "FiltrateExp",]
head(diversity_filtrate)
```

```r
# Test for streptomycin effect

diversity_filtrate$Streptomycin = as.factor(diversity_filtrate$Streptomycin)
M4 = lmer(Diversity ~ Streptomycin + Time_point + (1 | Replicate/Time_point), data = diversity_filtrate)
M5 = lmer(Diversity ~ Time_point + (1 | Replicate/Time_point), data = diversity_filtrate)
M6 = lmer(Diversity ~ Streptomycin + (1 | Replicate/Time_point), data = diversity_filtrate)
anova(M4, M5) # significant streptomycin effect
anova(M4, M6) # significant time effect
```

```r
# Post hoc test for streptomycin effect

summary(glht(M4, linfct = mcp(Streptomycin = 'Tukey')))
# Diversity differs between all concentrations
```

```r
# Plot

diversity_filtrate_summary = summarySE(diversity_filtrate, 
                                       measurevar = "Diversity", 
                                       groupvars = c("Streptomycin",
                                                     "Time_point"))
ggplot(diversity_filtrate_summary, aes(Time_point, Diversity, colour = Streptomycin)) + 
    geom_line() +
    geom_ribbon(aes(ymin = Diversity-se, ymax = Diversity+se, fill = Streptomycin), alpha = 0.5, colour = NA) +
    scale_colour_manual(values = mypalette[-3]) +
    scale_fill_manual(values = mypalette[-3]) +
    xlab("Time (h)") +
    ylab("Diversity (inverse Simpson index)") +
    theme_classic()

ggsave("../../../manuscript/figures/diversity_regime_shift_filtrate.pdf")
```

##### Summary for filtrate experiment: Diversity decreases with increasing streptomycin concentration (qualitatively or quantitatively better resource exploitation in communities subjected to streptomycin prior to filtrate collection) and over time.

## Repeatability of community trajectories (regime shift without predation)
### Source: https://www.biorxiv.org/content/early/2019/03/08/571505.full.pdf

```r
# Compute diversity (Shannon entropy)

diversity = specnumber(species_composition)
diversity = data.frame(Diversity = diversity, metadata)
diversity_regime_shift = diversity[diversity$Experiment != "FiltrateExp",]
diversity_regime_shift_nop = diversity_regime_shift[diversity_regime_shift$Predation == 0,]
div_ind = diversity_regime_shift_nop
head(div_ind)
```

```r
mean_div_ind = aggregate(Diversity ~ Streptomycin + Immigration + Time_point, data = div_ind, mean)
colnames(mean_div_ind) = c("Streptomycin", "Immigration", "Time_point", "Mean_diversity")
head(mean_div_ind)
```

```r
species_composition$ID = rownames(species_composition)
metadata$ID = rownames(metadata)
combined_data = merge(species_composition, metadata, by = "ID", all = T)
combined_data = combined_data[combined_data$Experiment == "RegimeShift" & combined_data$Predation == 0,]
head(combined_data)
```

```r
combined_data = combined_data[,-c(1, 31, 33, 36)]
head(combined_data)
```

```r
across_rep = aggregate(. ~ Streptomycin + Immigration + Time_point, data = combined_data, sum)
rownames(across_rep) = paste(across_rep$Streptomycin, across_rep$Immigration, across_rep$Time_point, sep = "_")
head(across_rep)
```

```r
across_rep_div = specnumber(across_rep[,-(1:3)])
head(across_rep_div)
```

```r
across_rep_div = data.frame(Pooled_diversity = across_rep_div, across_rep)
across_rep_div = across_rep_div[, 1:4]
rownames(across_rep_div) = c()
head(across_rep_div)
```

```r
div_all = merge(mean_div_ind, across_rep_div)
head(div_all)
```

```r
div_all$div_dissimilarity = (div_all$Pooled_diversity/div_all$Mean_diversity - 1)/(3-1)
head(div_all)
```

```r
# Shannon entropy

species_composition = read.table("../../../data/amplicon/species_composition_normalized_to_15k_reads.txt", 
                                 sep = "\t",
                                 header = T,
                                 row.names = 1)
diversity2 = diversity(species_composition, "shannon")
diversity2 = data.frame(Diversity = diversity2, metadata)
diversity_regime_shift2 = diversity2[diversity2$Experiment != "FiltrateExp",]
diversity_regime_shift_nop2 = diversity_regime_shift2[diversity_regime_shift2$Predation == 0,]
div_ind2 = diversity_regime_shift_nop2
mean_div_ind2 = aggregate(Diversity ~ Streptomycin + Immigration + Time_point, data = div_ind2, mean)
colnames(mean_div_ind2) = c("Streptomycin", "Immigration", "Time_point", "Mean_diversity")
across_rep_div2 = diversity(across_rep[,-(1:3)], "shannon")
across_rep_div2 = data.frame(Pooled_diversity = across_rep_div2, across_rep)
across_rep_div2 = across_rep_div2[, 1:4]
rownames(across_rep_div2) = c()
head(across_rep_div)
div_all2 = merge(mean_div_ind2, across_rep_div2)
div_all2$div_dissimilarity = (div_all2$Pooled_diversity/div_all2$Mean_diversity - 1)/(3-1)
```

```r
div_all$Type = "Species richness"
div_all2$Type = "Shannon entropy"
div_all3 = merge(div_all, div_all2, all = T)
div_all3
```

```r
# Add zero time point data for plotting

div_all3 = rbind(div_all3, c(0,0,0,0,0,0, "Species richness"))
div_all3 = rbind(div_all3, c(4,0,0,0,0,0, "Species richness"))
div_all3 = rbind(div_all3, c(16,0,0,0,0,0, "Species richness"))
div_all3 = rbind(div_all3, c(128,0,0,0,0,0, "Species richness"))
div_all3 = rbind(div_all3, c(0,1,0,0,0,0, "Species richness"))
div_all3 = rbind(div_all3, c(4,1,0,0,0,0, "Species richness"))
div_all3 = rbind(div_all3, c(16,1,0,0,0,0, "Species richness"))
div_all3 = rbind(div_all3, c(128,1,0,0,0,0, "Species richness"))

div_all3 = rbind(div_all3, c(0,0,0,0,0,0, "Shannon entropy"))
div_all3 = rbind(div_all3, c(4,0,0,0,0,0, "Shannon entropy"))
div_all3 = rbind(div_all3, c(16,0,0,0,0,0, "Shannon entropy"))
div_all3 = rbind(div_all3, c(128,0,0,0,0,0, "Shannon entropy"))
div_all3 = rbind(div_all3, c(0,1,0,0,0,0, "Shannon entropy"))
div_all3 = rbind(div_all3, c(4,1,0,0,0,0, "Shannon entropy"))
div_all3 = rbind(div_all3, c(16,1,0,0,0,0, "Shannon entropy"))
div_all3 = rbind(div_all3, c(128,1,0,0,0,0, "Shannon entropy"))
```

```r
div_all3$Immigration = as.factor(ifelse(div_all3$Immigration == 0, "Immigration absent", "Immigration present"))
```

```r
div_all3$div_dissimilarity = as.numeric(div_all3$div_dissimilarity)
div_all3$Time_point = as.numeric(div_all3$Time_point)
div_all3$Type = factor(div_all3$Type, levels = c("Species richness", "Shannon entropy"))
```

```r
# Plot

ggplot(div_all3, aes(Time_point, div_dissimilarity, colour = factor(Streptomycin))) +
    geom_line(size = 1) +
    facet_grid(.~Type*Immigration) +
    ylab("Diversity dissimilarity index") +
    theme_classic() +
    theme(strip.background = element_rect(colour = "white", fill = "white"),
          strip.text = element_text(face = "italic")) +
    scale_color_manual(values = c("#D3D3D3", "#cd6090", "#8f4364", "#522639")) + 
    xlab("Day") +
    labs(colour = "Streptomycin level") +
    scale_x_continuous(breaks = c(0, 16, 32, 48))
    #scale_y_continuous(limits = c(0, 0.5))

ggsave("../../../manuscript/figures/diversity_dissimilarity_index.pdf", width = 10)
```
