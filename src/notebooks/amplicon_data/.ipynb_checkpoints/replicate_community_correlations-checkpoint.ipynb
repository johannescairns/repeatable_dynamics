# Correlations between replicate communities

```r
# Housekeeping

library(ggplot2)
library(reshape2)
```

```r
# Read in data

species_composition = read.table("../../../data/amplicon/species_composition_relative_abundance.txt", 
                                 sep = "\t",
                                 header = T,
                                 row.names = 1)

metadata = read.table("../../../data/amplicon/metadata.txt",
                      sep = "\t",
                      header = T,
                      row.names = 1)

head(species_composition)
head(metadata)
```

```r
# Extract regime shift data without predation

x = metadata$Experiment == "RegimeShift" & # keep only regime shift data
    metadata$Predation == 0 & # exclude predation
    metadata$Immigration != "stock" # exclude initial strain mix data

# Subset

species_composition = species_composition[x,]
metadata = metadata[x,]

head(species_composition)
head(metadata)
```

```r
# Remove species lacking data

species_composition = species_composition[, colSums(species_composition) > 0]
dim(species_composition)
```

```r
# Merge and reshape data

species_composition$ID = rownames(species_composition)
metadata$ID = rownames(metadata)
df = merge(species_composition, metadata, all = T)
df = df[,-c(1, ncol(df), ncol(df)-3)]
df2 = melt(df, id.vars = c("Replicate", "Streptomycin", "Immigration", "Time_point"))
colnames(df2) = c("Replicate", "Streptomycin", "Immigration", "Time_point", "Species", "Abundance")
head(df)
head(df2)
```

```r
df3 = dcast(data = df2,formula = Immigration + Streptomycin + Species + Time_point ~ Replicate,fun.aggregate = sum,value.var = "Abundance")
colnames(df3) = c("Immigration", "Streptomycin", "Species", "Time_point", "REP1", "REP2", "REP3")
head(df3)
```

```r
ggplot(df3, aes(REP1, REP2, colour = Species)) + 
    geom_point() +
    geom_point(aes(y = REP3)) +
    scale_y_continuous(sec.axis = sec_axis(~., name = "REP3"), limits = c(min(df3[,5:7]), max(df3[,5:7]))) +
    scale_x_continuous(limits = c(min(df3[,5:7]), max(df3[,5:7]))) +
    guides(colour = guide_legend(ncol = 1)) +
    facet_grid(Immigration*Time_point ~ Streptomycin) +
    geom_smooth(method='lm',formula=y~x, colour = "grey", size = 0.5) +
    theme_classic() +
    theme(strip.background = element_rect(colour = "white", fill = "white"),
          strip.text = element_text(face = "italic"))
```

```r
# Create a correlation matrix

require(plyr)
func <- function(xx)
{
return(data.frame(COR1vs2 = cor(xx$REP1, xx$REP2), COR1vs3 = cor(xx$REP1, xx$REP3), COR2vs3 = cor(xx$REP2, xx$REP3)))
}

df4 = ddply(df3, .(Streptomycin, Immigration, Species), func)
df4
```

```r
# Inspect cases of particularly low correlation

df5 = na.omit(melt(df4, id.vars = c("Streptomycin", "Immigration", "Species")))
colnames(df5) = c("Streptomycin", "Immigration", "Species", "Comparison", "Correlation")
head(df5)
```

```r
ggplot(df5, aes(Correlation)) + geom_histogram(bins = 100) + theme_classic()

# The vast majority of species abundances are near-perfectly correlated
# between replicate communities per treatment combination (N = 3)
```

```r
# Let's examine correlations below 0

df6 = df5[df5$Correlation < 0,]
df6 = df6[order(df6$Correlation),]
head(df6)
```

```r
# Inspect species

x = sort(table(df6$Species), decreasing = T)
x[x > 0]

# i.e. most drastic deviations in species abundance between replicate communities
# are likely to be explained by these 15 species

# Since C. testosteroni frequently displays variability between replicated communities
# the visually noticeable deviation in replicate 2 from the
# 128 ug/ml streptomycin / no immigration treatment is unlikely to be caused by rapid evolution,
# unless the species displays recurrent rapid evolution, and variability in its abundance is more likely
# to be driven by species interactions
```
