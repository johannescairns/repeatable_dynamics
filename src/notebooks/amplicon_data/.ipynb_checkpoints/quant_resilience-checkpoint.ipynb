# Quantification of resilience

https://www.sciencedirect.com/science/article/pii/S0169534718300296

```r
# Housekeeping

library(ggplot2)
library(reshape2)
library(scales)
library(vegan)
source("../source.R")
```

```r
# Read in data

species_composition = read.table("../../../data/amplicon/species_composition_normalized_to_15k_reads.txt", 
                                 sep = "\t",
                                 header = T,
                                 row.names = 1)

metadata = read.table("../../../data/amplicon/metadata.txt",
                      sep = "\t",
                      header = T,
                      row.names = 1)

head(species_composition)
head(metadata)
```

```r
# Compute diversity (Shannon index)

diversity = diversity(species_composition, "shannon")
head(diversity)
```

```r
# Merge with metadata

diversity = data.frame(Diversity = diversity, metadata)
head(diversity)
```

```r
# Subset regime shift data without predation

# Select data (rows) to keep

diversity_regime_shift = diversity[diversity$Experiment != "FiltrateExp" &
                                   diversity$Predation == 0,]
head(diversity_regime_shift)
```

```r
div2 = dcast(diversity_regime_shift[,-c(4,7)], 
             Replicate + Streptomycin + Immigration ~ Time_point, 
             value.var = "Diversity")

head(div2)
```

```r
# Add impact data

div2$rec0 = div2$'32'/div2$'16'
div2$rec16 = div2$'48'/div2$'16'

div2
```

```r
# Edit for plotting

div3 = melt(div2[,-c(4:6)], id.vars = c("Replicate", "Streptomycin", "Immigration"))
colnames(div3) = c("Replicate", "Streptomycin", "Immigration", "Recovery_time", "System_state")
div3$Recovery_time = as.numeric(as.character(ifelse(div3$Recovery_time == "rec0", 0, 16)))
div3$System_state = ifelse(div3$System_state > 1, 1, div3$System_state)
head(div3)
```

```r
div4 = summarySE(div3, measurevar = "System_state", groupvars = c("Streptomycin", "Immigration", "Recovery_time"))
head(div4)
```

```r
div4$Immigration = as.factor(ifelse(div4$Immigration == 0, "Immigration absent", "Immigration present"))
```

```r
ggplot(div4, aes(Recovery_time, System_state, colour = factor(Streptomycin), fill = factor(Streptomycin))) +
    facet_grid(~Immigration) +
    geom_line() +
    geom_ribbon(aes(ymin = System_state-se, ymax = System_state+se), alpha = 0.5, colour = NA) +
    theme_classic() +
    theme(strip.background = element_rect(colour = "white", fill = "white"),
          strip.text = element_text(face = "italic")) +
    scale_color_manual(values = c("#D3D3D3", "#cd6090", "#8f4364", "#522639")) + 
    scale_fill_manual(values = c("#D3D3D3", "#cd6090", "#8f4364", "#522639")) + 
    xlab("Recovery time (days)") +
    ylab("System state") +
    scale_x_continuous(breaks = c(0,16)) +
    labs(colour = "Streptomycin level", fill = "Streptomycin level")

ggsave("../../../manuscript/figures/quantified_resilience.pdf")
```

```r
# Dissimilarity

D = as.matrix(vegdist(species_composition[metadata$Experiment != "FiltrateExp" &
                                   metadata$Predation == 0,]))
```

```r
meta2 = metadata[metadata$Experiment != "FiltrateExp" &
                                   metadata$Predation == 0,]
meta2$ID = rownames(meta2)
meta3 = dcast(meta2[,-c(3,6)], 
             Replicate + Streptomycin + Immigration ~ Time_point, 
             value.var = "ID")
head(meta3)
```

```r
meta3$rec0 = NA
meta3$rec16 = NA
head(meta3)
```

```r
for(i in 1:nrow(meta3)){
    a = meta3[i,4]
    b = meta3[i, 5]
    c = meta3[i, 6]
    meta3[i, "rec0"] = 1-D[rownames(D) == a, colnames(D) == b]
    meta3[i, "rec16"] = 1-D[rownames(D) == a, colnames(D) == c]
}
```

```r
# Edit for plotting

diss3 = melt(meta3[,-c(4:6)], id.vars = c("Replicate", "Streptomycin", "Immigration"))
colnames(diss3) = c("Replicate", "Streptomycin", "Immigration", "Recovery_time", "System_state")
diss3$Recovery_time = as.numeric(as.character(ifelse(diss3$Recovery_time == "rec0", 0, 16)))
diss3$System_state = ifelse(diss3$System_state > 1, 1, diss3$System_state)
head(diss3)
```

```r
diss4 = summarySE(diss3, measurevar = "System_state", groupvars = c("Streptomycin", "Immigration", "Recovery_time"))
head(diss4)
```

```r
diss4$Immigration = as.factor(ifelse(diss4$Immigration == 0, "Immigration absent", "Immigration present"))
```

```r
div4$type = "Shannon_diversity"
diss4$type = "Bray-Curtis_dissimilarity"
data = rbind(div4, diss4)
head(data)
```

```r
data$type = gsub("_", " ", data$type)
```

```r
ggplot(data, aes(Recovery_time, System_state, colour = factor(Streptomycin), fill = factor(Streptomycin))) +
    facet_grid(type~Immigration) +
    geom_line() +
    geom_ribbon(aes(ymin = System_state-se, ymax = System_state+se), alpha = 0.5, colour = NA) +
    theme_classic() +
    theme(strip.background = element_rect(colour = "white", fill = "white"),
          strip.text = element_text(size = 15, face = "italic"),
          axis.title.x = element_text(size = 15),
          axis.text.x = element_text(size = 12),
          axis.title.y = element_text(size = 15),
          axis.text.y = element_text(size = 12)) +
    scale_color_manual(values = c("#D3D3D3", "#cd6090", "#8f4364", "#522639")) + 
    scale_fill_manual(values = c("#D3D3D3", "#cd6090", "#8f4364", "#522639")) + 
    xlab("Recovery time (days)") +
    ylab("System state") +
    scale_x_continuous(breaks = c(0,16)) +
    labs(colour = "Streptomycin level", fill = "Streptomycin level")
```
