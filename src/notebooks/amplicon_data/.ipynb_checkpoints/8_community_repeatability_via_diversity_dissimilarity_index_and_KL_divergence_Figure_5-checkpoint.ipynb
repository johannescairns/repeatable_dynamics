# Quantifying repeatability of community trajectories using diversity dissimilarity index

Reference: https://www.biorxiv.org/content/10.1101/571505v1.full

**Housekeeping**

```r
library(boot)
library(ComplexHeatmap)
library(ggplot2)
library(multipanelfigure)
library(philentropy)
library(reshape2)
library(vegan)
```

**Read in data**

```r
species_composition = read.table("../../../data/amplicon/rel_abund.txt", 
                                 sep = "\t",
                                 header = T,
                                 row.names = 1)

metadata = read.table("../../../data/amplicon/meta.txt",
                      sep = "\t",
                      header = T,
                      row.names = 1)

# inspect

head(species_composition)
head(metadata)
```

**Compute repeatability for species richness**

Compute species richness for each community

```r
richness = specnumber(species_composition)
richness = data.frame(Diversity = richness, metadata)

# inspect

head(richness)
```

Compute mean species richness (across replicates) in each subtreatment and bootstrapped standard error

```r
# create identifier for each subtreatment (across replicates)

richness$Treatment = paste(richness$Streptomycin, richness$Immigration, richness$Day)
treatment = unique(richness$Treatment)

# script to loop over subtreatments and compute mean and standard error

mean_richness_list = list()

for(i in 1:length(treatment)){
    a = richness[richness$Treatment == treatment[i],]
    b = mean(a$Diversity) # compute mean diversity
    # wrapper script for bootstrapping
    foo = function(data, indices){
    dt = data[indices,]
    mean(dt[,"Diversity"])
    }
    set.seed(12345) # set seed for reproducible results
    # bootstrap standard error for mean species richness
    myBootstrap = boot(a, foo, R = 1000) 
    # create one-line data frame
    df = as.data.frame(a[1,])
    df = df[,-1]
    # add mean richness and standard error
    df$Richness_mean = b
    df$Richness_mean_se = sd(myBootstrap$t) # extract standard error from bootstrapping output
    # add to list file
    mean_richness_list[[i]] = df
}

# merge list into one data frame

mean_richness = do.call("rbind", mean_richness_list)
head(mean_richness)
```

Compute pooled species richness (across replicates) in each subtreatment and bootstrapped confidence intervals

_Pool all species together from replicate communities in each subtreatment_

```r
# combine species composition data with metadata

data = cbind(species_composition, metadata)

# remove column with replicate information

x = colnames(data) != "Replicate"
data = data[,x]

# pool species abundance data across replicates

pooled_abundance = aggregate(. ~ Streptomycin + Immigration + Day, data = data, sum)
rownames(pooled_abundance) = paste(pooled_abundance$Streptomycin, 
                                   pooled_abundance$Immigration, 
                                   pooled_abundance$Day, sep = "_")

# inspect

head(pooled_abundance)
```

Compute pooled species richness (across replicates) in each subtreatment

```r
treatment = unique(rownames(pooled_abundance))

pooled_richness_list = list()

for(i in 1:length(treatment)){
    a = as.data.frame(pooled_abundance[rownames(pooled_abundance) == treatment[i],])
    b = specnumber(a[,-(1:3)])
    c = a[,-c(1:3)]
    c = as.data.frame(t(ifelse(c > 0, 1, 0)))
    df = a[,c(1:3)]
    df$Richness_pooled = b
    pooled_richness_list[[i]] = df
}

# merge list into one data frame

pooled_richness = do.call("rbind", pooled_richness_list)

# inspect

head(pooled_richness)
```

Merge mean and pooled species richness data

```r
richness = merge(mean_richness, pooled_richness, by = c("Streptomycin", "Immigration", "Day"), all = T)

# inspect

head(richness)
```

Compute diversity dissimilarity index

```r
richness$d_d_index = (richness$Richness_pooled / richness$Richness_mean - 1) / (8 - 1)

# inspect

head(richness)
```

Propagate standard error for diversity dissimilarity index

```r
richness$d_d_index_richness_se = ((richness$Richness_pooled / richness$Richness_mean - 1) / (8 - 1)) * (richness$Richness_mean_se/richness$Richness_mean)

# inspect

head(richness)
```

**Compute repeatability for Shannon index**

Compute Shannon index for each community

```r
shannon = diversity(species_composition, "shannon")
shannon = data.frame(Diversity = shannon, metadata)

# inspect

head(shannon)
```

Compute mean Shannon index (across replicates) in each subtreatment and bootstrapped standard error

```r
# create identifier for each subtreatment (across replicates)

shannon$Treatment = paste(shannon$Streptomycin, shannon$Immigration, shannon$Day)
treatment = unique(shannon$Treatment)

# script to loop over subtreatments and compute mean and standard error

mean_shannon_list = list()

for(i in 1:length(treatment)){
    a = shannon[shannon$Treatment == treatment[i],]
    b = mean(a$Diversity) # compute mean diversity
    # wrapper script for bootstrapping
    foo = function(data, indices){
    dt = data[indices,]
    mean(dt[,"Diversity"])
    }
    set.seed(12345) # set seed for reproducible results
    # bootstrap standard error for mean Shannon index
    myBootstrap = boot(a, foo, R = 1000)
    df = as.data.frame(a[1,])
    df = df[,-1]
    # add mean Shannon index and standard error
    df$Shannon_mean = b
    df$Shannon_mean_se = sd(myBootstrap$t)
    mean_shannon_list[[i]] = df
}

# merge list into one data frame

mean_shannon = do.call("rbind", mean_shannon_list)
head(mean_shannon)
```

Compute pooled Shannon index (across replicates) in each subtreatment

```r
treatment = unique(rownames(pooled_abundance))

pooled_shannon_list = list()

for(i in 1:length(treatment)){
    a = as.data.frame(pooled_abundance[rownames(pooled_abundance) == treatment[i],])
    b = diversity(a[,-c(1:3)], "shannon")
    df = a[,c(1:3)]
    df$Shannon_pooled = b
    pooled_shannon_list[[i]] = df
}

# merge list into one data frame

pooled_shannon = do.call("rbind", pooled_shannon_list)

# inspect

head(pooled_shannon)
```

Merge mean and pooled Shannon index data

```r
shannon = merge(mean_shannon, pooled_shannon, by = c("Streptomycin", "Immigration", "Day"), all = T)

# inspect

head(shannon)
```

Compute diversity dissimilarity index

```r
shannon$d_d_index = (shannon$Shannon_pooled / shannon$Shannon_mean - 1) / (8 - 1)

# inspect

head(shannon)
```

Propagate standard error for diversity dissimilarity index

```r
shannon$d_d_index_Shannon_se = ((shannon$Shannon_pooled / shannon$Shannon_mean - 1) / (8 - 1)) * (shannon$Shannon_mean_se/shannon$Shannon_mean)

# inspect

head(shannon)
```

**Merge species richness and Shannon index data**

```r
# name columns consistently

colnames(richness) = c("Streptomycin",
                       "Immigration",
                       "Day",
                       "Replicate",
                       "Treatment",
                       "Diversity_mean",
                       "Diversity_mean_se",
                       "Diversity_pooled",
                       "d_d_index",
                       "d_d_index_se")

colnames(shannon) = c("Streptomycin",
                      "Immigration",
                      "Day",
                      "Replicate",
                      "Treatment",
                      "Diversity_mean",
                      "Diversity_mean_se",
                      "Diversity_pooled",
                      "d_d_index",
                      "d_d_index_se")

# add diversity type column

richness$Type = "Species richness"
shannon$Type = "Shannon entropy"
diversity = merge(richness, shannon, all = T)

# subset

diversity = diversity[,c("Streptomycin", 
                         "Immigration",
                         "Day", 
                         "d_d_index",
                         "d_d_index_se",
                         "Type")]

# inspect

head(diversity)
```

**KL divergence**

Create KL distance matrix for all communities

```r
KL = KL(as.matrix(species_composition))
rownames(KL) = rownames(species_composition)
colnames(KL) = rownames(species_composition)

# inspect

head(KL)
```

Edit treatment information before creating annotation for upcoming heat map

```r
metadata$Day = ifelse(metadata$Day == 16, "Before pulse", 
                            ifelse(metadata$Day == 32, "During pulse", "After recovery"))

metadata$Streptomycin = ifelse(metadata$Streptomycin == 0, "No",
                              ifelse(metadata$Streptomycin == 4, "Low",
                                    ifelse(metadata$Streptomycin == 16, "Intermediate", "High")))

metadata$Immigration = ifelse(metadata$Immigration == 0, "No immigration", "Immigration")
```

Create annotation for heat map

```r
annot_df = data.frame(Immigration = metadata$Immigration, Antibiotic = metadata$Streptomycin, 
                       Time = metadata$Day)

# define colors for each levels of qualitative variables

col = list(Antibiotic = c("No" = "#D3D3D3", "Low" = "#cd6090", "Intermediate" = "#8f4364", "High" = "#522639"),
           Time = c("Before pulse" = "lightgrey", "During pulse" = "black", "After recovery" = "darkgrey"),
           Immigration = c("No immigration" = "yellow", "Immigration" = "orange"))

# create heat map annotation

ha = HeatmapAnnotation(annot_df, 
                       col = col,
                       annotation_legend_param = list(title_gp = gpar(fontsize = 16),
                                                      labels_gp = gpar(fontsize = 16)))
```

Create heat map

```r
p2 = Heatmap(KL,
             name = "KL divergence",
             top_annotation = ha,
             show_row_names = F,
             show_column_names = F,
             heatmap_legend_param = list(title_gp = gpar(fontsize = 16), 
                                         labels_gp = gpar(fontsize = 16)))
p2
```
