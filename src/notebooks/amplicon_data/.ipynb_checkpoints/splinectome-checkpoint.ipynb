```r
# Housekeeping

# First, load the required packages
library(dplyr)
library(tibble)
library(ggplot2)
library(reshape2)
library(tidyr)
#devtools::install_github('RRShieldsCutler/splinectomeR')
library(splinectomeR)
library(vegan)
```

```r
# Read in data

species_composition = read.table("../../../data/amplicon/species_composition_relative_abundance.txt", 
                                 sep = "\t",
                                 header = T,
                                 row.names = 1)

metadata = read.table("../../../data/amplicon/metadata.txt",
                      sep = "\t",
                      header = T,
                      row.names = 1)


# Extract regime shift data without predation

x = metadata$Experiment != "FiltrateExp" & # keep only regime shift data
    metadata$Predation != 1 # exclude predation

# Subset

species_composition = species_composition[x,]
metadata = metadata[x,]

head(species_composition)
head(metadata)
```

```r
unique(metadata$Time_point)
```

```r
# Merge data

species_composition$ID = rownames(species_composition)
metadata$ID = rownames(metadata)
df = merge(species_composition, metadata, all = T)
dim(df)
head(df)
```

```r
df2 = melt(df, id.vars = c("ID", "Replicate", "Streptomycin", "Predation", "Immigration", "Time_point", "Experiment"))
colnames(df2) = c("ID", "Replicate", "Streptomycin", "Predation", "Immigration", "Time_point", "Experiment", "Species", "Abundance")
head(df2)
```

```r
df2$Immigration = ifelse(df2$Immigration == "stock", 1, df2$Immigration)
unique(df2$Immigration)
head(df2)
```

```r
# Add stock data

df3 = df2[df2$Replicate != "stock",]
stock = df2[df2$Replicate == "stock",]
stock0 = stock
stock0$Immigration = 1
stock$Immigration = 2
stock0_0 = stock0
stock4_0 = stock0
stock4_0$Streptomycin = 4
stock16_0 = stock0
stock16_0$Streptomycin = 16
stock128_0 = stock0
stock128_0$Streptomycin = 128
stock0_1 = stock
stock4_1 = stock
stock4_1$Streptomycin = 4
stock16_1 = stock
stock16_1$Streptomycin = 16
stock128_1 = stock
stock128_1$Streptomycin = 128
df4 = rbind(stock0_0, stock4_0, stock16_0, stock128_0, stock0_1, stock4_1, stock16_1, stock128_1, df3)
head(df4)
```

```r
# New ID column

df4$Replicate = ifelse(df4$ID == "Im_stock1", 1, ifelse(df4$ID == "Im_stock2", 2, 3))
df4$ID = paste(df4$Streptomycin, df4$Immigration, df4$Replicate, sep = "_")
head(df4)
```

```r
# Remove redundant columns

df4 = df4[,-c(4, 7)]
head(df4)
```

```r
Aeromonas_caviae_HAMBI_1972_result <- permuspliner(data = df4[df4$Species == "Aeromonas_caviae_HAMBI_1972",],
                                                   xvar = 'Time_point',
                                                   yvar = 'Abundance', 
                                                   perms = 999,
                                                   category = "Immigration", 
                                                   #groups = c(0, 4), 
                                                   cases = 'ID', 
                                                   retain_perm = T,
                                                   set_tol = 0.1,
                                                   set_spar = 2)
```
